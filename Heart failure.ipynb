```python
required = ['anaemia','diabetes','high_blood_pressure','sex','smoking']
```

```python
df=pd.get_dummies(df,columns=required)
df
```

```python
for i in objects_columns:
    df[i+'_log'] = np.log(df[i]+1)
```

```python
df.head()
```

```python
df =df.drop(objects_columns,axis=1)
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
```

```python
from sklearn.preprocessing import LabelEncoder
label_en = LabelEncoder()
df['anaemia_0']= label_en.fit_transform(df['anaemia_0'])
df['anaemia_1']= label_en.fit_transform(df['anaemia_1'])
df['diabetes_0']= label_en.fit_transform(df['diabetes_0'])
df['diabetes_1']= label_en.fit_transform(df['diabetes_1'])
df['high_blood_pressure_0']= label_en.fit_transform(df['high_blood_pressure_0'])
df['high_blood_pressure_1']= label_en.fit_transform(df['high_blood_pressure_1'])
df['sex_0']= label_en.fit_transform(df['sex_0'])
df['sex_1']= label_en.fit_transform(df['sex_1'])
df['smoking_0']= label_en.fit_transform(df['smoking_0'])
df['smoking_1']= label_en.fit_transform(df['smoking_1'])
```

```python
X = df.drop('DEATH_EVENT',axis=1)
y = df['DEATH_EVENT']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,stratify = y)
```

```python
log_model = LogisticRegression()
log_model.fit(X_train,y_train)
```

```python
log_ypred = log_model.predict(X_test)
```

```python
log_acc = accuracy_score(y_test,log_ypred)
print(log_acc)
```

```python
log_clf = classification_report(y_test,log_ypred)
print(log_clf)
```

```python
from sklearn.svm import SVC
svc_model= SVC(kernel='linear')
svc_model.fit(X_train,y_train)
```

```python
svc_ypred = svc_model.predict(X_test)
```

```python
svc_acc = accuracy_score(y_test,svc_ypred)
print(svc_acc)
svc_clf = classification_report(y_test, svc_ypred)
print(svc_clf)
```

```python
from sklearn.tree import DecisionTreeClassifier
df_model = DecisionTreeClassifier()
df_model.fit(X_train,y_train)
```

```python
df_ypred = df_model.predict(X_test)
df_acc = accuracy_score(y_test,df_ypred)
print(df_acc)
```

```python
df_clf = classification_report(y_test,df_ypred)
print(df_clf)
```
